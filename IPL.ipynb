```python
data = pd.read_csv('cleaned_ipl_matches.csv')

print("First few rows of the dataset:")
print(data.head())
```

```python

print("\nSummary statistics:")
print(data.describe())
```

```python

print("\nShape of the dataset:", data.shape)

```

```python

plt.figure(figsize=(10, 6))
sns.countplot(y='city', data=data, order=data['city'].value_counts().index)
plt.title('Number of Matches Played in Each City')
plt.xlabel('Number of Matches')
plt.ylabel('City')
plt.show()
```

```python

plt.figure(figsize=(8, 5))
sns.countplot(x='result', data=data)
plt.title('Outcome of Matches')
plt.xlabel('Match Result')
plt.ylabel('Count')
plt.show()
```

```python

team_counts = pd.concat([data['team1'], data['team2']]).value_counts()
plt.figure(figsize=(10, 6))
sns.barplot(x=team_counts.values, y=team_counts.index, palette='viridis')
plt.title('Number of Matches Played by Each Team')
plt.xlabel('Number of Matches')
plt.ylabel('Team')
plt.show()
```

```python

numeric_data = data.select_dtypes(include=['float64', 'int64'])
numeric_data.drop('id', axis=1, inplace=True)  # Drop 'id' column
plt.figure(figsize=(10, 6))
sns.heatmap(numeric_data.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(data['result_margin'], bins=20, kde=True)
plt.title('Distribution of Match Margins')
plt.xlabel('Match Margin')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(x='toss_decision', data=data)
plt.title('Toss Decisions')
plt.xlabel('Toss Decision')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(y='umpire1', data=data, order=data['umpire1'].value_counts().index)
plt.title('Number of Matches Umpired by Umpire 1')
plt.xlabel('Number of Matches')
plt.ylabel('Umpire 1')
plt.show()
```

```python

data['date'] = pd.to_datetime(data['date'])

data['year'] = data['date'].dt.year

plt.figure(figsize=(10, 6))
sns.countplot(x='year', data=data)
plt.title('Number of Matches Played Each Year')
plt.xlabel('Year')
plt.ylabel('Number of Matches')
plt.show()
```

```python
print(data.dtypes)
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='year', hue='winner', data=data)
plt.title('Number of Matches Won by Each Team Each Year')
plt.xlabel('Year')
plt.ylabel('Number of Matches')
plt.legend(title='Winner', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.countplot(y='venue', hue='winner', data=data, order=data['venue'].value_counts().index[:10])
plt.title('Top 10 Venues by Number of Matches')
plt.xlabel('Number of Matches')
plt.ylabel('Venue')
plt.legend(title='Winner', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(y='player_of_match', data=data, order=data['player_of_match'].value_counts().index[:10])
plt.title('Top 10 Players of the Match')
plt.xlabel('Number of Awards')
plt.ylabel('Player')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(x='toss_decision', hue='winner', data=data)
plt.title('Toss Decision vs. Match Outcome')
plt.xlabel('Toss Decision')
plt.ylabel('Count')
plt.legend(title='Winner', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```
